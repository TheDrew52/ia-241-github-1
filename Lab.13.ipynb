```python
%matplotlib inline
import pandas
```

```python
df = pandas.read_excel('s3://ia241-1-spring2022-drew/covid_data.xls')
df[:39301]
```

```python
sum_cases = df.groupby('countriesAndTerritories').sum()['cases']
```

```python
sum_cases.plot()
```

```python
death_sum = df.groupby('countriesAndTerritories').sum()['deaths']
```

```python
death_sum.nlargest(10).plot.bar()
```
